```python
import pandas as pd
from simpletransformers.seq2seq import Seq2SeqModel,Seq2SeqArgs
```

```python
model_args=Seq2SeqArgs()
model_args.num_train_epochs=10 
model_args.no_save=True
model_args.evaluate_generated_text=True
model_args.evaluate_during_text=True
model_args.evaluate_during_traning_verbose=True

model = Seq2SeqModel(
           encoder_decoder_type="bart",
           encoder_decoder_name="facebook/bart-large",
           args=model_args,
           use_cuda=False,
)

```

```python
train_data=[
    [
        "Sophie found a lost puppy on her way back home from school. She took him home and decided to keep him as her pet.",
        "Sophie found a lost puppy and decided to keep him as her pet.",
    ],
   [
       "Samantha woke up early on her birthday to find her parents had surprised her with a trip to Disneyland. She was overjoyed and couldn't wait to go on all the rides and meet her favorite characters.",
       "Samantha's parents surprised her with a trip to Disneyland on her birthday. She was excited to go on rides and meet her favorite characters"
   ],
]
train_df=pd.DataFrame(train_data,columns=["input_text","target_text"])
```

```python
eval_data=[
    [
        "John had been saving his allowance for months to buy a new bike. When he finally had enough money, he went to the store and picked out the perfect one.",
        "John saved his allowance for months to buy a new bike, and when he finally had enough money, he went to the store and picked out his perfect bike.",
    ],
    [
        "Emily had always wanted to learn how to play the piano, so she signed up for lessons at the local music school. With practice and dedication, she became a skilled pianist.",
        "Emily signed up for piano lessons and with practice and dedication, she became a skilled pianist.",
    ],
]
eval_df=pd.DataFrame(eval_data,columns=["input_text","target_text"])
```

```python
train_df
```

```python
def count_matches(labels,preds):
  print(labels)
  print(preds)
  return sum(
      [
          1  if label==pred else 0
          for label,pred in zip(labels,preds)
      ]
  )
```

```python
model.train_model(train_df,eval_data=eval_df,matches=count_matches)
```

```python
results=model.eval_model(eval_df)
```

```python
results
```
